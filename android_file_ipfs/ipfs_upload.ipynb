```python
import ipfshttpclient
import os
from pathlib import Path
from web3 import Web3
import json

path=Path()
client=ipfshttpclient.connect('/ip4/127.0.0.1/tcp/5001/http')
print(client)
url = "http://localhost:7545"
web3 = Web3(Web3.HTTPProvider(url))
print(web3.isConnected())
web3.eth.defaultAccount = web3.eth.accounts[0]
address=web3.toChecksumAddress('0x632f1deeab3d144e2fe0a9f12919e493a9e52488')


abi = json.loads('[{"constant":true,"inputs":[],"name":"get_data","outputs":[{"name":"","type":"string"}],"payable":false,"stateMutability":"view","type":"function"},{"constant":false,"inputs":[{"name":"_name","type":"string"}],"name":"set_data","outputs":[],"payable":false,"stateMutability":"nonpayable","type":"function"},{"constant":true,"inputs":[],"name":"data","outputs":[{"name":"","type":"string"}],"payable":false,"stateMutability":"view","type":"function"}]')
contract = web3.eth.contract(address=address, abi=abi)
```

```python
while True:
    f=open('status.txt','r')
    status=f.read()
    f.close()
    if status=='upload':
        f=open('fname.txt','r')
        fname=f.read()
        f.close()
        print("fname",fname)
        res=client.add(fname)
        print(res)
        print(res['Name'])
        print(res['Hash'])
        print(res['Size'])
        os.remove(fname)
        hash_fname=fname+"_hash.txt"
        print("read",hash_fname)
        f=open(hash_fname,'r')
        fhash=f.read()
        f.close()
        print("fhash",fhash)
        blk_info=fname+"##@@"+fhash+"##@@"+res['Hash']+'##@@'+res['Size']
        print(blk_info)

        chain_data=contract.functions.get_data().call()
        print("old data",chain_data)
        new_data=chain_data+"%%@@"+blk_info
        resp = contract.functions.set_data(new_data).transact()
        web3.eth.waitForTransactionReceipt(resp)
        print('resp=',resp)
        f=open('status.txt','w')
        f.write('')
        f.close()
        os.remove(hash_fname)
    elif status=='get_file_names':
        f=open('user.txt','r')
        email=f.read()
        f.close()
        chain_data=contract.functions.get_data().call()
        print("old data",chain_data)
        l=chain_data.split('%%@@')
        f=open('list.txt','w')
        f.write('')
        f.close()
        f=open("list.txt",'a')
        for x in l:
            l1=x.split('##@@')
            if str(l1[0]).startswith(email):
                arr=l1[0].split('_')
                fname=arr[1]
                fhash=l1[1]
                ipfs_hash=l1[2]
                fsize=l1[3]
                print(fname,fhash,ipfs_hash,fsize)
                f.write(fname+"##"+fhash+"##"+ipfs_hash+"##"+fsize+"\n")
        f.close()
        f=open('status.txt','w')
        f.write('')
        f.close()
    elif status=='get_file':
        f=open('fname.txt','r')
        fname=f.read()
        f.close()
        f=open('list.txt','r')
        fdet=f.read()
        f.close()
        l=fdet.split('\n')
        print("l=",l)
        for x in l:
            print(x,type(x))
            l1=x.split('##')
            if l1[0]==fname:
                fhash=l1[1]
                ipfs_hash=l1[2]
                getfile=client.cat(ipfs_hash)
                f = open(fname, "wb")
                f.write(getfile)
                f.close()
        f=open(fname+'_hash.txt','w')
        f.write(fhash)
        f.close()
        
        f=open('status.txt','w')
        f.write('')
        f.close()
        
        
        
        
        
        
        
            
            
                
                
                
                
                
                
```
